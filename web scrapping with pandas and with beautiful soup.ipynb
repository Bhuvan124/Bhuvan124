```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url = ("https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue")
page = requests.get(url)
soup = BeautifulSoup(page.text,"html")
print(soup)
```

```python
soup.find('table')
```

```python
soup.find_all("table")[0]
```

```python
soup.find("table",class_ = "wikitable sortable")
```

```python
soup.find_all("table")[1]
```

```python
table = soup.find_all("table")[1]
print(table)
```

```python
soup.find_all("th")
```

# we need to extarct data from the tables not from the soup

```python
world_titles = table.find_all("th")
```

```python
world_titles
```

# list comprehesnions 

```python
world_title = [data.text.strip() for data in world_titles]
```

```python
print(world_title)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_title)
df
```

```python
column_data = table.find_all("tr")
```

```python
column_data
```

```python
column_data = table.find_all("tr")
```

```python
for row in column_data[1:]:
    each_data=row.find_all("td")
    each_row_values = [data.text.strip() for data in each_data] ## we are taking each row from t
df.loc[len(df)] = each_row_values
```

```python
df
```

```python
df.info()
```

```python
df["Revenue (USD millions)"] = pd.to_numeric(df["Revenue (USD millions)"].replace(',', ''), errors='coerce')
df["Revenue growth"] = pd.to_numeric(df["Revenue growth"].replace('%', ''), errors='coerce')
df["Employees"] = pd.to_numeric(df["Employees"].replace(',', ''), errors='coerce')
```

```python
import seaborn  as sns
```

```python
graphs1 = sns.barplot(x="Revenue growth",y = "Industry" ,data=df)
```

```python
graphs3 = sns.barplot(x="Revenue (USD millions)",y = "Industry" ,data=df)
```
